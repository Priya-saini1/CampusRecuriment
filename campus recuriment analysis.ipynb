# objective of Analysis

# question1:find out how many students have been placed or not in different streams  

```python
df.head()
```

```python
sns.countplot(x="status",hue="degree_t",data=df,palette="Set1",linewidth=3,edgecolor=sns.color_palette("dark", 2))
sns.set_theme(style="darkgrid")
plt.xlabel("Status")
plt.ylabel("Number of candidate")
plt.title("No of students placed or not placed in different streams")
plt.show()
```

# question2 which degree specialization is much demand by corporate?

```python
sns.countplot(x="status",hue="specialisation",data=df,palette="muted",linewidth=3,edgecolor=sns.color_palette("magma", 2))
sns.set_theme(style="darkgrid")
plt.xlabel("Status")
plt.ylabel("Number of candidate")
plt.title("Specialization demanded by corporate")
plt.show()
```

# question 3 does work experience affects placement of a candiate 


```python
sns.countplot(x="status",hue="workex",data=df,palette="dark",linewidth=3,edgecolor=sns.color_palette("pastel", 2))
sns.set_theme(style="darkgrid")
plt.xlabel("Status")
plt.ylabel("Number of candidate")
plt.title("Candidate Work Experience")
plt.show()
```

# question 4 find out how many male and female have been placed or not 

```python
sns.countplot(x="status",hue="gender",palette="bright",data=df,linewidth=3,edgecolor=sns.color_palette("magma", 2))
sns.set_theme(style="darkgrid")
plt.xlabel("Gender")
plt.ylabel("Number of candidate")
plt.title("How many Male and Female got placed ")
plt.show()
```

```python
###find does gender of an students affects the salary or any othervariable 
sns.boxplot(x=df['gender'],y=df['salary'],hue='specialisation',data=df)
```
